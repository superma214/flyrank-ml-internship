## Research Question

**Decision**

Which pages should be reviewed first for a content refresh?

**Unit of analysis**

A single webpage.

**Output**

A ranked list of pages with a refresh opportunity score.

**Action**

The SEO or content team reviews the highest-ranked pages first and decides whether to update them.

**Cost of a wrong recommendation**

If pages are ranked incorrectly, time may be spent reviewing pages that do not need updates while important declining pages may be missed. This reduces the efficiency of the content review process.

**Why ML can help**

Multiple signals such as impressions, CTR, average position, content age, and update history interact together. Machine learning can combine these signals better than a simple hand-written rule, providing useful decision support.

## Quick Look at the Data

The starter dataset contains approximately **30,000 pages**, providing enough examples for exploratory analysis.

The correlation between **search volume and impressions is approximately 0.001**, suggesting that search volume alone is not a strong indicator of page impressions.

The median word count is similar across different trend categories, indicating that content length alone is unlikely to explain page performance.

These observations suggest that multiple features should be considered together when prioritizing pages for content refresh.

```python
import os
import sys
import subprocess

REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if not os.path.isdir(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

os.chdir(REPO_DIR)

print("Current directory:", os.getcwd())
print("Dataset exists:", os.path.exists("data/raw/content_refresh_anonymized.csv"))

import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", len(df))

print()

corr = df["search_volume"].corr(df["impressions_90d"])
print("Correlation between search volume and impressions:", round(corr,3))

print()

print("Median word count by trend:")
print(df.groupby("trend_direction")["word_count"].median())

print()

print("Number of declining pages:")
print((df["trend_direction"]=="down").sum())
```
